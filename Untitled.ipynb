```python

def set_seed(sd=123):
    from numpy.random import seed

    import random as rn
    ## numpy random seed
    seed(sd)
    ## core python's random number 
    rn.seed(sd)
    ## tensor flow's random number
    tf.random.set_seed(sd)
```

# GENERATE DATA

```python
def random_sample(len_ts=3000,D=1001):
    c_range = range(5,100)
    c1 = np.random.choice(c_range)
    u = np.random.random(1)
    const = -1.0/len_ts
    ts = np.arange(0,len_ts)
    
    x1 = np.cos(ts/float(1.0 + c1))
    x1 = x1*ts*u*const
    
    y1 = np.zeros(len_ts)

    for t in range(D,len_ts):
        ## the output time series depend on input as follows: 
        y1[t] = x1[t-2]*x1[t-D] 
    y = np.array([y1]).T
    X = np.array([x1]).T
    return y, X
def generate_data(D= 1001,Nsequence = 1000,T=4000, seed=123):
    X_train = []
    y_train = []
    set_seed(sd=seed)
    for isequence in range(Nsequence):
        y, X = random_sample(T,D=D)
        X_train.append(X)
        y_train.append(y)
    return np.array(X_train),np.array(y_train)

D = 10
T = 1000
X, y = generate_data(D=D,T=T,Nsequence = 1000)
print(X.shape, y.shape)
```

# PLOT DATA

```python
def plot_examples(X,y,ypreds=None,nm_ypreds=None):
    fig = plt.figure(figsize=(16,10))
    fig.subplots_adjust(hspace = 0.32, wspace = 0.15)
    count = 1
    n_ts = 16
    for irow in range(n_ts):
        ax = fig.add_subplot(n_ts/4,4,count)
        ax.set_ylim(-0.5,0.5)
        ax.plot(X[irow,:,0],"--",label="x1")
        ax.plot(y[irow,:,:],label="y",linewidth=3,alpha = 0.5)
        ax.set_title("{:}th time series sample".format(irow))
        if ypreds is not None:
            for ypred,nm in zip(ypreds,nm_ypreds):
                ax.plot(ypred[irow,:,:],label=nm)   
        count += 1
    plt.legend()
    plt.show()
plot_examples(X,y,ypreds=None,nm_ypreds=None)
```

# AR MODEL

```python
def ARMODEL(x, a, sigma, c=0):
    assert len(x) == len(a)
    X = c + sigma
    for i in range(len(x)):
        X += x[i]*a[i]
    return X
```

```python
# Generate
N = 1000
X0 = []
a1 = [0.7,-0.04,0.05,0.1]
x0 = [0.3368983609609942, 0.3467158354287806, 0.37901651148323984, 0.36451171310125796]

for i in range(N):
    sigma = random.random()/10
    r = ARMODEL(x0, a1, sigma)
    x0 = [r] + x0[:3]
    X0.append(r)
```

# MA MODEL

```python
import numpy as np
def MAMODEL(x, a, sigma, c=0):
    
    assert len(x) == len(a)
    
    X = np.mean(x) 
    
    for i in range(len(x)):
        
        X += sigma[i]*a[i]
        
    return X
```

```python
# Generate
N = 1000
X2 = []
a1 = [0.1,-0.1,-0.2,0.2]
x0 = [1,1,1,1]

for i in range(N):
    sigma = [random.random() for x in range(len(a1))]
    r = MAMODEL(x0, a1, sigma)
    x0 = [r] + x0[:3]
    X2.append(r)
```

```python
X2
plt.figure(figsize=(16,10))

plt.plot(range(N),X2,label="y", linewidth=3, alpha = 0.5)
plt.show()
```

# Generate 2D data 4x4


```python
#   x0->x0->x0->x0->
#   x1->x1->x1->x1->
#   x2->x2->x2->x2->
#   x3->x3->x3->x3->

def ARMODEL(x, a, sigma, c=0):
    assert len(x) == len(a)
    X = c + sigma
    for i in range(len(x)):
        X += x[i]*a[i]
    return X

def MAMODEL(x, a, sigma, c=0):
    assert len(x) == len(a)
    X = np.mean(x) 
    for i in range(len(x)):   
        X += sigma[i]*a[i]  
    return X

```

```python
#X1
N = 1000
a1 = [0.7,-0.04,0.05,0.1]
X1 = [0.3368983609609942, 0.3467158354287806, 0.37901651148323984, 0.36451171310125796]

for i in range(N):
    sigma = (random.random()-0.5)/100
    x1 = ARMODEL(X1[-4:], a1, sigma)
    X1.append(x1)
X1 = [(np.sin(x/50)+1)/100 for x in range(N+4)]
```

```python
#X2
def X1MODEL(x0, x, a, sigma, c=0):
    assert len(x) == len(a)
    L = len(x)
    X = np.mean(x) 
    for i in range(L):   
        X += sigma[i]*a[i]  
    X = X[:L] + x0
    return X

N = 1000
# X0 already created 
X2 = [0.3806983609609942, 0.3942158354287806, 0.39841651148323984, 0.37961171310125796] # for start
a2 = [0.7, -0.4, 0.05, 0.1] # weights in ARMODEL
b2 = [0.2, 0.8] # weights in MAMODEL
mu0 = -0.01 # mu
mu1 = -0.2 # real mu
# x0 already created

for i in range(N):
    sigma = (random.random()-0.5)/100
    
    x2 = mu0 + b2[0]*(X1[i]-mu1) + b2[1]*(ARMODEL(X2[-4:], a2, sigma)-mu1) + sigma # MA part
    X2.append(x2)
```

```python
plt.plot(np.array([X1,X2]).T)

plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title(
    "Realizations of Geometric Brownian Motion with different variances\n $\mu=1$"
)
plt.show()
```
